```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
%time iso_map = gpd.read_file(".\Independent_System_Operator_9131146937303417597.zip")
iso_map.crs
```

### Convert Geographic CRS to Projected CRS

```python
%time iso_map.to_crs(epsg=5070, inplace=True)
iso_map.crs
```

```python
fig, ax = plt.subplots(figsize=(6,6))
iso_map.iloc[:1,:].to_crs(epsg=4326).plot(ax=ax,edgecolor='k', color='gray')
iso_map.iloc[:1, -1].centroid.to_crs(epsg=4326).plot(ax=ax)
```
